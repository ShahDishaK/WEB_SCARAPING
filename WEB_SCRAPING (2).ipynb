```python
import pandas as pd
# website = 'https://www.politifact.com/factchecks'
# result = requests.get(website)
# content = result.text
# soup = BeautifulSoup(content , 'html.parser')
```

```python
# ul = soup.find('ul' , {'class' : 'o-listicle__list'})
# li = ul.find_all('li' , {'class' : 'o-listicle__item'})
l=[]
frame = []
user = int(input('Enter Page NUmber : '))
for outer in range(1 , user+1):
    website = f'https://www.politifact.com/factchecks/list/?page={outer}'
    result = requests.get(website)
    content = result.text
    soup = BeautifulSoup(content , 'html.parser')
    ul = soup.find('ul' , {'class' : 'o-listicle__list'})
    li = ul.find_all('li' , {'class' : 'o-listicle__item'})
    for i in li:
        div_quote = i.find('div' , {'class' : 'm-statement__quote'}).text.strip()
        a = i.find('div' , {'class' : 'm-statement__quote'}).find('a')['href']
        link = 'https://www.politifact.com'
        link+=a
        date = i.find('footer').text.split('.')[-1]
        source = i.find('div' , class_ = 'm-statement__meta').text.strip()
        meter = i.find('div' , class_='m-statement__meter')
        label = meter.find('img' , class_ = 'c-image__original')['alt']
        frame.append((div_quote , link , date , source , label) )
    
data = pd.DataFrame(frame , columns = ['statement' , 'link' , 'date' , 'source' , 'label'])
data
```

```python
f = open('Scrape_this.html' , 'r')
soup = BeautifulSoup(f , 'html.parser')
table = soup.find('table' , {'class':'common-table'}).find_all('tr')[1:]
df=[]
# print(table)
for i in table:
    d={}
    d['Company Name'] = i.find('td' , {'class' : ""}).find('a')['title']
    d['Company link'] = i.find('td' , {'class' : ""}).find('a')['href']
    d['Current Price'] = i.find_all('td')[1].get_text()
    d['Value (₹ Cr.)'] = i.find_all('td')[5].get_text()
    d['Volume (in 000s)'] = i.find_all('td')[6].get_text()
    df.append(d)

data = pd.DataFrame(df)
data.head()
```

```python
data = pd.DataFrame(df , columns=['Current Price' ,'Value (₹ Cr.)'] , dtype=float)
data.isna().sum()
```
